#### 3. 수치형 변수에 대하여 최소-최대 표준화 하기 : 사용자 정의 함수 활용

```python
# method : select_dtypes(), apply()
# function : sklearn.preprocessing.minmax_scale()
```

```python
[(value - df["CO2배출량"].mean())/( df["CO2배출량"].max() - df["CO2배출량"].min()) for value in df["CO2배출량"]][0:10]
```

```python
for column in df1.select_dtypes(include="number").columns:
    df1[column+"_minmax"] = [(value - df1[column].mean())/( df1[column].max() - df1[column].min()) for value in df1[column]]
```

```python
# 데이터 분할 : train, test
from sklearn.model_selection import train_test_split
# 선형회귀 분석
from sklearn.linear_model import LinearRegression 
# 모델성능 평가
from sklearn.metrics import r2_score, mean_squared_error 
```

```python
# import sklearn error 발생, python version 변경으로 해결
```

#### 1. 단순선형회귀분석

#### 1) 모델링

```python
plt.scatter(df["배기량"], df["CO2배출량"])
plt.show()
```

```python
train, test = train_test_split(df, train_size=0.7)
```

```python
df.shape, train.shape, test.shape
```

```python
lr=LinearRegression()
```

```python
type(lr)
```

```python
lr.fit(train[["배기량"]], train["CO2배출량"])    # x : dataframe 
```

```python
vars(lr)
```

#### 2) 예측

```python
y_pred=lr.predict(test[["배기량"]])
y_pred
```

```python
plt.scatter(test["배기량"], test["CO2배출량"])
plt.scatter(test["배기량"], y_pred)
plt.show()
```

```python
plt.scatter(test["CO2배출량"], y_pred)
plt.xlabel("actual")
plt.ylabel("predicted value")
```

#### 3) 잔차 분석

```python
residual = test["CO2배출량"] - y_pred
```

```python
plt.hist(residual)
plt.show()
```

#### 4) 모델 성능 평가 

```python
r2_score(test["CO2배출량"], y_pred)
```
